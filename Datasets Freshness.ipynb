```python
import numpy as np
import pandas as pd
import datetime
from datetime import datetime,date
import urllib
 
from bokeh.plotting import *
from bokeh.models import HoverTool
from collections import OrderedDict
```

```python
# Read in our data. We've aggregated it by date already, so we don't need to worry about paging
query = ("https://data.lacity.org/resource/kqy8-vgbp.json?")
raw_data = pd.read_json(query)
```

```python
raw_data.head()
```

```python
raw_data.drop(['publication_stage', 'public','routing_approval', 'source_link','derived_view',
'type', 'what_geographic_unit_is_the_data_collected', 'does_this_data_have_a_location_column_yes_or_no', 'domain', 'data_provided_by','owner_uid'], axis=1, inplace=True)
```

```python
raw_data.head()
```

```python
for i in range(0,len(raw_data)):
    raw_data['last_update_date_data'][i]=datetime.utcfromtimestamp(raw_data['last_update_date_data'][i]).strftime('%Y-%m-%d')
    
print ("Done converting date")
```

```python
for i in range(0,len(raw_data)):
    raw_data['creation_date'][i]=datetime.utcfromtimestamp(raw_data['creation_date'][i]).strftime('%Y-%m-%d')
    
print ("Done converting date")
```

```python
raw_data.tail()
```

```python
raw_data = raw_data[['name','description','department','category','license','owner','contactemail','creation_date','last_update_date_data', 'visits','downloads', 'dataset_link','api_endpoint','refresh_rate']]
```

```python
raw_data.tail()
```

```python
raw_data.groupby('refresh_rate').size()
```

```python
raw_data['is_dataset_up_to_date']=''
raw_data['action_needed']=''
```

```python
now = datetime.now()
```

```python
raw_data['days_since_last_update']=0
for i in range(0,len(raw_data)):
    td=now-datetime.strptime(raw_data['last_update_date_data'][i], "%Y-%m-%d")
    raw_data['days_since_last_update'][i]=td.days
    
print ('Done calculating days diff')
```

```python
raw_data.head()
```

```python
for i in range(0,len(raw_data)):
    if  pd.isnull(raw_data['refresh_rate'][i]):
        raw_data['action_needed'][i]='Contact dataset owner to update Refresh rate'
        raw_data['is_dataset_up_to_date'][i]='No'
    elif raw_data['refresh_rate'][i]=='Annual':
        if raw_data['days_since_last_update'][i]<375:
            raw_data['action_needed'][i]= 'No action needed'
            raw_data['is_dataset_up_to_date'][i]='Yes'
        else:
            raw_data['action_needed'][i]= 'Contact dataset owner to update dataset'
            raw_data['is_dataset_up_to_date'][i]='No'
    elif raw_data['refresh_rate'][i]=='Daily':
        if raw_data['days_since_last_update'][i]<3:
            raw_data['action_needed'][i]= 'No action needed'
            raw_data['is_dataset_up_to_date'][i]='Yes'
        else:
            raw_data['action_needed'][i]= 'Contact dataset owner to update dataset'
            raw_data['is_dataset_up_to_date'][i]='No'
    elif raw_data['refresh_rate'][i]=='Hourly':
        if raw_data['days_since_last_update'][i]<2:
            raw_data['action_needed'][i]= 'No action needed'
            raw_data['is_dataset_up_to_date'][i]='Yes'
        else:
            raw_data['action_needed'][i]= 'Contact dataset owner to update dataset'
            raw_data['is_dataset_up_to_date'][i]='No'
    elif raw_data['refresh_rate'][i]=='Monthly':
        if raw_data['days_since_last_update'][i]<34:
            raw_data['action_needed'][i]= 'No action needed'
            raw_data['is_dataset_up_to_date'][i]='Yes'
        else:
            raw_data['action_needed'][i]= 'Contact dataset owner to update dataset'
            raw_data['is_dataset_up_to_date'][i]='No'
    elif raw_data['refresh_rate'][i]=='Quarterly':
        if raw_data['days_since_last_update'][i]<140:
            raw_data['action_needed'][i]= 'No action needed'
            raw_data['is_dataset_up_to_date'][i]='Yes'
        else:
            raw_data['action_needed'][i]= 'Contact dataset owner to update dataset'
            raw_data['is_dataset_up_to_date'][i]='No'
    elif raw_data['refresh_rate'][i]=='Semi-annual':
        if raw_data['days_since_last_update'][i]<190:
            raw_data['action_needed'][i]= 'No action needed'
            raw_data['is_dataset_up_to_date'][i]='Yes'
        else:
            raw_data['action_needed'][i]= 'Contact dataset owner to update dataset'
            raw_data['is_dataset_up_to_date'][i]='No'
    elif raw_data['refresh_rate'][i]=='Weekly':
        if raw_data['days_since_last_update'][i]<10:
            raw_data['action_needed'][i]= 'No action needed'
            raw_data['is_dataset_up_to_date'][i]='Yes'
        else:
            raw_data['action_needed'][i]= 'Contact dataset owner to update dataset'
            raw_data['is_dataset_up_to_date'][i]='No'
    else:
        raw_data['action_needed'][i]= 'No action needed'
        raw_data['is_dataset_up_to_date'][i]='Yes'
        
        
print ('Done setting logic')
        
            
            
        
    
```

```python
raw_data.drop('days_since_last_update', inplace=True, axis=1)
```

```python
raw_data.rename(columns={'name':'Dataset Name',
                          'description':'Description',
                          'department':'Department',
                            'license': 'License',
                            'owner': 'Owner',
                            'contactemail': 'Contact Email',
                            'creation_date': 'Creation Date',
                            'last_update_date_data': 'Last Update Date',
                        'visits': 'Visits',
                        'downloads': 'Downloads',
                        'dataset_link': 'Dataset Link',
                        'api_endpoint' : 'API Endpoint',
                        'refresh_rate': 'Refresh Rate',
                        'is_dataset_up_to_date': 'Is Dataset Up to Date',
                        'action_needed': 'Action Needed'}, 
                 inplace=True)
```
